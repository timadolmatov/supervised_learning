```python
#УДАЛЕНИЕ ИЗ ДАТАСЕТА НЕНУЖНЫХ СТОЛБЦОВ RowNumber, CustomerId, Surname
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
display(data)
```

```python
#ПРЯМОЕ КОДИРОВАНИЕ
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score




data_OHE = pd.get_dummies(data, drop_first=True)
display(data_OHE)
display(data_OHE.corr())
target_OHE = data_OHE['Exited']
features_OHE = data_OHE.drop('Exited', axis=1)

#КОВАРИАЦИОННАЯ МАТРИЦА
display(features_OHE.corr())

features_train_OHE, features_other_OHE, target_train_OHE, target_other_OHE = train_test_split(features_OHE, target_OHE, 
                                test_size=0.4, random_state=12345) # < напишите код здесь >
features_valid_OHE, features_test_OHE, target_valid_OHE, target_test_OHE = train_test_split(features_other_OHE, 
                                target_other_OHE, test_size=0.5, random_state=12345)

#МАСШТАБИРОВАНИЕ ПРИЗНАКОВ

numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
scaler = StandardScaler()
scaler.fit(features_train_OHE[numeric])
features_train_OHE[numeric] = scaler.transform(features_train_OHE[numeric])
display(features_train_OHE.corr())
features_valid_OHE[numeric] = scaler.transform(features_valid_OHE[numeric])
display(features_valid_OHE.corr())
features_test_OHE[numeric] = scaler.transform(features_valid_OHE[numeric])
display(features_test_OHE.corr())



#ОБУЧЕНИЕ МОДЕЛИ ЛОГИСТИЧЕСКОЙ РЕГРЕССИИ
model = LogisticRegression(random_state=12345) # инициализируйте модель логистической регрессии с параметром random_state=12345
model.fit(features_train_OHE, target_train_OHE) # обучите модель на тренировочной выборке
print("Обучено!")

pred_valid_OHE = model.predict(features_valid_OHE)
f_1 = f1_score(pred_valid_OHE, target_valid_OHE)
print('F1-мера базовой модели равна:', f_1)

#АДЕКВАТНОСТЬ МОДЕЛИ

result_OHE = model.score(features_valid_OHE, target_valid_OHE) # получите метрику качества модели на валидационной выборке
print("Accuracy модели на валидационной выборке:", result_OHE)
class_frequency_OHE = target_valid_OHE.value_counts(normalize=True)# < напишите код здесь >)
print(class_frequency_OHE)
class_frequency_OHE.plot(kind='bar')

```

```python
#УВЕЛИЧЕНИЕ ВЫБОРКИ
from sklearn.utils import shuffle

def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train_OHE, target_train_OHE, 3)

model_2 = LogisticRegression(random_state=12345)
model_2.fit(features_upsampled, target_upsampled)# < напишите код здесь >
predicted_valid_OHE = model_2.predict(features_valid_OHE)

print("F1:", f1_score(target_valid_OHE, predicted_valid_OHE))
```

```python
#УМЕНЬШЕНИЕ ВЫБОРКИ
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train_OHE, target_train_OHE, 0.4)

model_3 = LogisticRegression(random_state=12345)
model_3.fit(features_downsampled, target_downsampled)
predicted_valid = model_3.predict(features_valid_OHE)# < напишите код здесь >

print("F1:", f1_score(target_valid_OHE, predicted_valid))
```
